```python
    x = np.arange(-2*np.pi, 2*np.pi, 0.1)
    print(x)

    y_sin = np.sin(x)
    y_cos = np.cos(x)
    y_tan = np.tan(x)
    y_arctan = np.arctan(x)

    plt.subplot(2, 2, 1)
    plt.plot(x, y_sin)
    plt.title('Sine')

    plt.subplot(2, 2, 2)
    plt.plot(x, y_cos)
    plt.title('Cosine')

    plt.subplot(2, 2, 3)
    plt.plot(x, y_tan)
    plt.title('Tan')

    plt.subplot(2, 2, 4)
    plt.plot(x, y_arctan)
    plt.title('Arttan')
    plt.show()
```

在数学中，正弦（英语：sine、缩写 {\displaystyle \sin } \sin）是一种周期函数，是三角函数的一种。它的定义域是整个实数集，值域是 {\displaystyle [-1,1]} [-1,1]。它是周期函数，其最小正周期为 {\displaystyle 2\pi } 2\pi 。在自变量为 {\displaystyle {\frac {(4n+1)\pi }{2}}} {\displaystyle {\frac {(4n+1)\pi }{2}}}（ {\displaystyle n} n为整数）时，该函数有极大值1；在自变量为 {\displaystyle {\frac {(4n+3)\pi }{2}}} {\displaystyle {\frac {(4n+3)\pi }{2}}}时，该函数有极小值-1。正弦函数是奇函数，其图像于原点对称。

```python

n=1
x = ((4*n + 1)*np.pi) / 2

np.sin(x)
```
